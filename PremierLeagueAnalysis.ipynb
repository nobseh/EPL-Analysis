# Generate a Pivot Table for All Referees sorted by Season

```python
# Create a Pivot Table 'pvt' and sort the values in descending order
pvt = pd.pivot_table(data, index = "Referee", values = "Season", aggfunc = "count").sort_values(by = "Season", ascending= False)
pvt
```
